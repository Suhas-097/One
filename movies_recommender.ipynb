```python
# recommendation model
```

```python
movie_user=Master_data.pivot(index='UserID',columns='Movie_Name',values='Rating')
```

```python
movie_user.head()
```

```python
# Showing the movies with strong co relation to "Toy Story (1995)"
```

```python
correlations=movie_user.corrwith(movie_user['Toy Story (1995)'])
correlations.head()
```

```python
recommendation=pd.DataFrame(correlations,columns=['Correlation'])
```

```python
recommendation.dropna(inplace=True)
```

```python
#Calculating the average rating of each movie
```

```python
Average_ratings=pd.DataFrame(Master_data.groupby('Movie_Name')['Rating'].mean())
```

```python
Average_ratings['Total Ratings']=pd.DataFrame(Master_data.groupby('Movie_Name')['Rating'].count())
```

```python
recommendation=recommendation.join(Average_ratings['Total Ratings'])
```

```python
recommendation.head()
```

```python
recommended_movies=recommendation[recommendation['Total Ratings']>100].sort_values('Correlation',ascending=False).reset_index()
```

```python
recommended_movies=recommended_movies.merge(movie,on='Movie_Name',how='left')
```
